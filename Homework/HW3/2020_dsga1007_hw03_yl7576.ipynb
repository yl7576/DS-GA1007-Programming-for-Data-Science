### Problem 01 (20 points)

(10 points) Plot at the same diagram the functions: $1, x, x^2, x^3, \dots, x^{10}$ over the following set of points $\{0, 0.01, 0.02, \dots, 1\}$, that is, over 101 equidistant points from the interval $[0, 1]$. 


(10 points) Plot at the same diagram the unctions:f $1, x, x^{1/2}, x^{1/3}, \dots, x^{1/10}$ over the following set of points $\{0, 0.01, 0.02, \dots, 1\}$, that is, over 101 equidistant points from the interval $[0, 1]$. 




```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

    
```

```python
# Question 1
points = np.linspace(0,1,101)
lst=[]
for j in range(0,11):
    lst.append(points**j)
    
dt = pd.DataFrame(lst)
dt = dt.transpose()
dt.index = points

plt.figure(figsize=(10,8))
plt.plot(dt)
plt.xlabel('points')
plt.legend(range(0,11), title = 'exponential to x', )

#plt.ylim((0,10))
```

```python
# Question 2
exp = []
for i in range(1,11):
    exp.append(1/i)
exp.append(0)

lst2 = []
for j in exp:
    lst2.append(points**j)
```

```python
dt2 = pd.DataFrame(lst2)
dt2 = dt2.transpose()
dt2.index = points
dt2.columns = exp
dt2

plt.figure(figsize=(10,8))
plt.plot(dt2)
plt.xlabel('points')
plt.legend(exp, title = 'exponential to x', )
```
