### Problem 02 (30 points)

2.1.1 (10 points) Create a pandas dataframe with the indexes $0, 1, \dots, 9$ and column labels "col0", "col1", ..., "col9". Each value of the dateframe with the index $i$ and with the column "colj" should be assigned a random value from the uniform distribution on $[0, 1]$. Hint: One can use, as in the class, https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html . 


2.1.2 (5 points) Filter the dataframe from 2.1.1 so to remove all rows for which the corresponding value in the column "col7" is smaller than 0.2 and bigger than 0.8

2.2.1 (10 points) Create a pandas dataframe with the indexes $0, 1, \dots, 9$ and column labels "col0", "col1", ..., "col9". Each value of the dateframe with the index $i$ and with the column "colj" should be assigned the value $50 \cdot i + j$. 

2.2.2 (5 points) Filter the dataframe from 2.2.1 so to remove all rows for which the corresponding value in the column "col7" is smaller than 100 and bigger than 350


### Solution 02 (30 points)

````python
# 2.1.1
colnames = ['col0','col1','col2','col3','col4','col5','col6','col7','col8','col9']
df = pd.DataFrame(np.random.uniform(0, 1, (10, 10)), columns=colnames)
df
````

````python
# 2.1.2
df1 = df[df.col7>=0.2][df.col7<=0.8]
df1
````

````python
# 2.2.1
df2 = pd.DataFrame(index = range(0,10), columns=colnames)
for i in range(0,10):
    for j in range(0,10):
        df2.iloc[[i],[j]] = 50*i+j
df2
````

````python
# 2.2.2
df3 = df2[df2.col7>=100][df2.col7<=350]
df3
````
